```python
import pandas as pd
```

```python
%%capture
df = pd.read_csv('data/_tmp_transactions.csv.gz', compression='gzip')
```

```python
df
```

```python
df_eip1559 = df[df['type'] == "0x02"]
df_eip1559
```

```python
%%capture
df_eip1559['max_fee_gwei'] = df_eip1559['max_fee'].astype(float) / 1e9
df_eip1559['max_priority_fee_gwei'] = df_eip1559['max_priority_fee'].astype(float) / 1e9
```

```python
df_eip1559
```

```python
ax = df_eip1559.plot(x='block_number', y='max_fee_gwei', kind='scatter', color='C0')
df_eip1559.plot(x='block_number', y='max_priority_fee_gwei', kind='scatter', ax=ax, color='C1')
```

```python
# proportion of txs in a block that are 1559
df_grouped_by_type = df.groupby(by='block_number').agg({'type': lambda x: x[x == "0x02"].count() / x.count() })
```

```python
df_grouped_by_type = df_grouped_by_type.drop(['block_number'])
df_grouped_by_type['number'] = df_grouped_by_type.index
df_grouped_by_type
```

```python
df_grouped_by_type.plot(x='number', y='type', kind='scatter', label='fraction of 1559 txs in block')
```
